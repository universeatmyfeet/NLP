```
import pandas as pd
```

```
db = pd.read_json('http://snap.stanford.edu/data/amazon/productGraph/categoryFiles/reviews_Office_Products_5.json.gz',lines = True)#reading gz file from url
db
```

```
db['reviewText'] = db['reviewText'].str.lower()#all characters become lower-cased
db['summary'] = db['summary'].str.lower()
```

```
def counting_words(text):
  word_list = text.split() #returns list of strings with characters as elements
  total_words = len(text.split())# total words in the string
  # Initializing Dictionary
  d = {'total_count' : total_words}
  # counting number of times each word comes up in list of words (in dictionary)
  for word in word_list: 
     d[word] = d.get(word, 0) + 1 # adding entry to the dictionary , with key as 'word' and value : (default value for initial entry to dictionary = 0) + 1 = 1 
     # value gets incremented if the word occurs again
  return d
```

```
db['reviewText_wordcount'] = db.apply(lambda row : counting_words(row['reviewText']), axis = 1) 
# apply() function in order to apply count_words function to every row of the column
db['summary_wordcount'] = db.apply(lambda row : counting_words(row['summary']), axis = 1) 
```
